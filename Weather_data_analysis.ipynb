```python
import pandas as pd

def get_season(date):
    """Converts a date to its corresponding season."""

    if isinstance(date, str):
        date = pd.to_datetime(date)

    month = date.month

    if 3 <= month <= 5:
        return "Spring"
    elif 6 <= month <= 8:
        return "Summer"
    elif 9 <= month <= 11:
        return "Autumn"
    else:
        return "Winter"

weather_df["season"] = weather_df["Date_Time"].apply(get_season)
display(weather_df)
```
