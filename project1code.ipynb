```python
weather_df=pd.read_csv("weather_data.csv")
```

```python
weather_df.head()
```

```python
weather_df["Date_Time"] 
```

```python
weather_df["Date_Time"]=pd.to_datetime(weather_df["Date_Time"])
```

```python
weather_df.head()
```

```python
def get_season(date):
    """Converts a date to its corresponding season."""

    if isinstance(date, str):
        date = pd.to_datetime(date)

    month = date.month

    if 3 <= month <= 5:
        return "Spring"
    elif 6 <= month <= 8:
        return "Summer"
    elif 9 <= month <= 11:
        return "Autumn"
    else:
        return "Winter"

weather_df["season"] = weather_df["Date_Time"].apply(get_season)
display(weather_df)
```

```python
location_df = weather_df.groupby(["Location","season"])
display(weather_df)
```
